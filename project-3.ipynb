```python
import re 
import requests
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
import numpy as np
```

```python
driver = webdriver.Chrome(executable_path='chromedriver/chromedriver')
driver.get('https://www.noon.com/saudi-en/electronics-and-mobiles/computers-and-accessories/laptops?limit=200')
sleep(2) 
html = driver.page_source 
driver.close()
```

```python
response = BeautifulSoup(html,'html.parser')
```

```python
response.status_code
```

```python
s=response.find_all('p',class_="jsx-599068390 name")
```

```python
len(s)
```

```python
laptop_name = []
for i in range(0,200):
    laptop_name.append(s[i].string)
    
laptop_name
```

```python
price=response.find_all("span" ,class_="jsx-4251264678 sellingPrice")
```

```python
price
```

```python
laptop_price= []
for i in price:
    laptop_price.append(i.findChildren('span',class_="value")[0].text)
    
len(laptop_price)
```

```python
len(laptop_price)
```

```python
df=pd.DataFrame()
df['Laptop_Describtion']=laptop_name
df['laptop_price']=laptop_price
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 150)
```

```python
df.head()
```

```python
b=response.find_all("p" ,class_="jsx-599068390 brand")
len(b)
```

```python
brand_name = []
for i in range(0,200):
    brand_name.append(b[i].string)
brand_name    
```

```python
df['Brand_name']=brand_name
df.head()
```

```python
de=response.find_all("span" ,class_="jsx-2551926997 discountTag")
len(de)
```

```python
d=response.find_all("div" ,class_="jsx-599068390 detailsContainer")
len(d)
```

```python
s=df.Laptop_Describtion.apply(lambda  x : str(x).split(',')[-1])

```

```python
c=s.apply(lambda  x : str(x).split('/')[0])
```

```python
df['Core']=c
df.head()
```

```python
df.Laptop_Describtion.str.extract('(\d{1,3}) ?GB') + 'GB'
```

```python
y=df.Laptop_Describtion.apply(lambda  x : str(x).split(',')[0])
y
```

```python
df['Inch']=y.apply(lambda  x : str(x).split('With')[-1].split('/')[0].split(',')[0])
```
